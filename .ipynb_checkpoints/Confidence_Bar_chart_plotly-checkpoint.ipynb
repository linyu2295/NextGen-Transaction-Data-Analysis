# Plotly - Bar Chart


```python
'''
    df: confidence matrix
    category: 'restaurant', 'grocery'

'''

def plotly_bar_chart(df, category):
    # reorder the columns by name
    df = df.sort_index(axis = 1, ascending=False)

    # for each feature: Descending order of yaxis

    fig = go.Figure()

    buttons = []

    for row in df.index.to_list():
        fig.add_trace(
            go.Bar(
                x = df.loc[row][np.argsort(df.loc[row])].values,
                y = df.loc[row][np.argsort(df.loc[row])].index,
    #             x = df.loc[row],
    #             y = df.columns,
                name = row,
                orientation = 'h')) # set the trace name, which appears as the legend item and on hover

        buttons.append(dict(method = 'update',
                           label = row,
                           visible = True,
                           args = [{'visible': df.index.isin([row]),
                                   'title': row,
                                    'type': 'bar',
                                   'showlegend': True}]))
    fig.update_layout(autosize=False,
#                       xaxis = dict(showgrid=True, 
#                                    domain = [0, 1]), 
                      title = '{}: The confidence value of the rule X->Y'.format(category.capitalize()),
                      xaxis_title = 'Confidence: P(Y|X)',
                      yaxis_title = 'Feature Y',
                      legend_title = 'Feature X',
                    width=1200,
                    height=800,
                    barmode = 'stack',
                    updatemenus = [go.layout.Updatemenu(
                        active = 0,
                        direction = 'down',
                        yanchor = 'top',
                        y = 1.05,
                        buttons = buttons)]
                                 )

    fig.show()    
    
    return fig
```

```python
fig_restaurant = plotly_bar_chart(conf_mat_restaurant, category = 'restaurant')
```

```python
fig_grocery = plotly_bar_chart(conf_mat_grocery, category = 'grocery')
```
